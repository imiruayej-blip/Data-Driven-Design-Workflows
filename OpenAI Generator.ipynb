```python
from openai import OpenAI
import base64

client = OpenAI()

prompt = """
Create a clean fashion illustration of a model wearing editorial streetwear.
Use crisp outlines, flat colors, minimal shading, white background,
and a polished magazine-sketch look.
"""

response = client.responses.create(
    model="gpt-5.5",
    input=prompt,
    tools=[{"type": "image_generation"}],
)

image_base64 = None

for item in response.output:
    if item.type == "image_generation_call":
        image_base64 = item.result
        break

with open("test_fashion.png", "wb") as f:
    f.write(base64.b64decode(image_base64))

print("saved test_fashion.png")
```

```python
import base64
from openai import OpenAI

client = OpenAI()

def encode_image(path):
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

image_path = "harpers_image_0.jpg"
base64_image = encode_image(image_path)

prompt = """
Convert this photo into a clean fashion illustration.
Preserve the pose and clothing silhouette, but simplify details into crisp outlines,
flat colors, minimal shading, and a white background.
"""

response = client.responses.create(
    model="gpt-4.1-mini",
    input=[
        {
            "role": "user",
            "content": [
                {"type": "input_text", "text": prompt},
                {"type": "input_image", "image_url": f"data:image/jpeg;base64,{base64_image}"},
            ],
        }
    ],
    tools=[{"type": "image_generation"}],
)

image_base64 = None
for item in response.output:
    if item.type == "image_generation_call":
        image_base64 = item.result
        break

with open("fashion_illustration.png", "wb") as f:
    f.write(base64.b64decode(image_base64))

print("saved fashion_illustration.png")
```

```python
from pathlib import Path
from openai import OpenAI
import base64
import mimetypes

client = OpenAI()

def encode_image(path):
    with open(path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")

image_path = Path.home() / "Downloads" / "harpersbazaar_images" / "harpers_image_0.jpg"

print("Using:", image_path)
print("Exists:", image_path.exists())

mime_type, _ = mimetypes.guess_type(str(image_path))
if mime_type is None:
    mime_type = "image/jpeg"

base64_image = encode_image(str(image_path))

prompt = """
Convert this photo into a clean fashion illustration.
Preserve the pose and outfit silhouette, but simplify details into crisp outlines,
flat colors, minimal shading, and a clean white background.
Make it look like a polished editorial sketch for a fashion magazine.
"""

response = client.responses.create(
    model="gpt-4.1-mini",
    input=[
        {
            "role": "user",
            "content": [
                {"type": "input_text", "text": prompt},
                {"type": "input_image", "image_url": f"data:{mime_type};base64,{base64_image}"},
            ],
        }
    ],
    tools=[{"type": "image_generation"}],
)

image_base64 = None
for item in response.output:
    if item.type == "image_generation_call":
        image_base64 = item.result
        break

if image_base64 is None:
    raise ValueError("No image was returned. Check the API response and model/tool support.")

output_path = Path.home() / "Downloads" / "harpersbazaar_images" / "harpers_image_0_illustrated.png"

with open(output_path, "wb") as f:
    f.write(base64.b64decode(image_base64))

print("Saved to:", output_path)
```
