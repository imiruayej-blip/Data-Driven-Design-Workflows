```python
import pandas as pd
import json
import matplotlib.pyplot as plt

with open("/Users/jitminaj/Downloads/harpersbazaar_images/designers.json") as f:
    data = json.load(f)

df = pd.DataFrame(data)

# Filter out unknowns
df = df[df["designer"] != "unknown"]

# Plot top 10 designers
df["designer"].value_counts().head(10).plot(kind="barh", figsize=(10, 6), color="black")

plt.title("Most Featured Designers - Met Gala 2026")
plt.xlabel("Number of Looks")
plt.tight_layout()
plt.show()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load both JSON files
with open("/Users/jitminaj/Downloads/vogue_images/designers.json") as f:
    vogue = json.load(f)

with open("/Users/jitminaj/Downloads/harpersbazaar_images/designers.json") as f:
    hb = json.load(f)

hb_clean = [{"celebrity": x["celebrity"], "designer": x["designer"], "caption": x["caption"]} for x in hb]

all_data = vogue + hb_clean
df = pd.DataFrame(all_data)
df = df[df["celebrity"] != "unknown"]

# Count celebrity appearances
celebrity_counts = df["celebrity"].value_counts().head(20)

# Plot
fig, ax = plt.subplots(figsize=(12, 8))
celebrity_counts.plot(kind="barh", ax=ax, color="#457b9d")

ax.set_title("Most Featured Celebrities - Met Gala 2026", fontsize=16)
ax.set_xlabel("Number of Appearances")
ax.set_ylabel("Celebrity")
ax.invert_yaxis()
plt.tight_layout()
plt.savefig("/Users/jitminaj/Downloads/celebrity_frequency.jpg", dpi=300, bbox_inches="tight")
plt.show()

print("Saved to Downloads/celebrity_frequency.jpg")
```
